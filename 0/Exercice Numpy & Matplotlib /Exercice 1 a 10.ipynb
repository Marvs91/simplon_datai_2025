Tracer un graphe linéaire des points (x, y) où x = [0, 1, 2, 3, 4] et y = [1, 4, 3, 8, 5].

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = [0, 1, 2, 3, 4]
y = [1, 4, 3, 8, 5]

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title('Graphe Linéaire X et Y')
plt.grid(True)
plt.show()
```

Tracer un histogramme à partir de données aléatoires générées par NumPy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Données aléatoires
np.random.seed(100)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# Histogramme
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogramme')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```

Tracer un diagramme à barres à partir des données suivantes :

categories = ['A', 'B', 'C', 'D', 'E'] valeurs = [25, 30, 15, 20, 10]

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
valeurs = [25, 30, 15, 20, 10]

plt.bar(categories, valeurs)  # plt.bar() pour créer un diagramme à barres
plt.xlabel("Catégories")
plt.ylabel("Valeurs")
plt.title("Diagramme à barres")
plt.show()
```

Tracer un nuage de points à partir de données aléatoires générées par NumPy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Données aléatoires
np.random.seed(42) 
x = np.random.rand(50)  # 50 valeurs aléatoires pour l'axe x
y = np.random.rand(50)  # 50 valeurs aléatoires pour l'axe y

# Nuage de points
plt.scatter(x, y)

# Etiquettes et un titre
plt.xlabel("Axe X")
plt.ylabel("Axe Y")
plt.title("Nuage de points aléatoires")

# Le graphique
plt.show()


```

Tracer une courbe sinusoïdale en utilisant NumPy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Données pour la courbe sinusoïdale
x = np.linspace(0, 2 * np.pi, 100)  # 100 points répartis entre 0 et 2*pi
y = np.sin(x)  # Calcul de la valeurs de sin(x)

# Graphique de la courbe sinusoïdale
plt.plot(x, y)

# Étiquettes et un titre
plt.xlabel("Angle (radians)")
plt.ylabel("sin(x)")
plt.title("Courbe sinusoïdale")

# Le graphique
plt.show()
```

Tracer deux courbes sur le même graphe à partir des données suivantes :

x = np.linspace(0, 10, 100) y1 = np.sin(x) y2 = np.cos(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 300)
y1 = np.sin(x)
y2 = np.cos(x)

# Graphique avec les deux courbes
plt.plot(x, y1)  # Première courbe avec étiquette 'sin(x)'
plt.plot(x, y2)  # Deuxième courbe avec étiquette 'cos(x)'

# Étiquettes d'axe et un titre
plt.xlabel("x")
plt.ylabel("y")
plt.title("Courbes de sin(x) et cos(x)")

# Le graphique
plt.show()
```

Tracer un diagramme en secteurs (camembert) à partir des données suivantes : categories = ['A', 'B', 'C', 'D'], valeurs = [30, 25, 15, 20]

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
valeurs = [30, 25, 15, 20]

# Le diagramme en secteurs (camembert)
plt.pie(valeurs, labels=categories, autopct='%1.1f%%', startangle=100)

# Le titre
plt.title("Diagramme en secteurs (Camembert)")

# Le graphique
plt.show()
```

Tracer une courbe polynomiale à partir des données suivantes :

    x = np.linspace(-5, 5, 100)
    y = 3 * x**3 + 2 * x**2 - 8 * x + 5

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

# Le graphique de la courbe polynomiale
plt.plot(x, y)

# Étiquettes d'axe et un titre
plt.xlabel("x")
plt.ylabel("y")
plt.title("Courbe polynomiale")

# Le graphique
plt.show()
```

Tracer un diagramme en barres empilées à partir des données suivantes :

categories = ['A', 'B', 'C', 'D'] valeurs1 = [10, 15, 20, 25] valeurs2 = [5, 10, 15, 20]

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
valeurs1 = [10, 15, 20, 25]
valeurs2 = [5, 10, 15, 20]

# Le graphique en barres empilées
plt.bar(categories, valeurs1)
plt.bar(categories, valeurs2)

# Étiquettes d'axe et un titre
plt.xlabel("Catégories")
plt.ylabel("Valeurs")
plt.title("Diagramme en barres empilées")

# Le graphique
plt.show()
```

Tracer un graphique en boîte (boxplot) à partir des données suivantes :

data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(-3, 2, 100)]

```python
import numpy as np
import matplotlib.pyplot as plt

data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(-3, 2, 100)]

# Le graphique en boîte
plt.boxplot(data, labels=['Groupe 1', 'Groupe 2', 'Groupe 3'])

# Le titre
plt.title("Graphique en boîte")

# Le graphique
plt.show()
```
